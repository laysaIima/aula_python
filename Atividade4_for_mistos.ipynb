# Exercícios C — for

## 1. Mostre os números de 1 até 5.

```python
for numero in range(6):
    print(numero, end=',')
```

## 2. Mostre os números pares de 0 até 20.

```python
for numero in range(0, 20, 2):
    print(numero, end=',')
```

## Mostre a tabuada do 5.

```python
for numero in range(1, 11):
    print(f"5x {numero} =", 5*numero)
```

## 4. Repita a frase "Python é incrível" 4 vezes.

## 5. Percorra a palavra "Pernambuco" e mostre cada letra.

```python
for letra in "Pernambuco":
    print(letra)
```

## 6. Some os números de 1 até 10 usando for.

```python
soma = 0
for numero in range(1, 11):
    soma = soma + numero

print(f"A soma dos números de 1 até 10 é: {soma}")
```

# Exercícios D — Mistos

## 1. Peça 5 números com for e diga quantos são pares.

```python
pares = 0
for i in range(1, 6):
    numero = int(input(f"Digite o {i}º número: "))
    if numero % 2 == 0:
        pares = pares + 1

print(f"\nQuantidade de números pares: {pares}")
```

## 2. Use while para pedir uma nota entre 0 e 10.

```python
nota = -1

while nota < 0  or nota > 10:
    nota = float(input("Digite uma nota entre 0 e 10: "))
    if nota < 0 or nota > 10:
        print("Nota inválida! Tente novamente.")

print(f"Nota aceita: {nota}")
```

## 3. Peça um número e mostre a tabuada com for.

```python
numero = int(input("Digite um número: "))

for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

## 4. Peça idade e documento. Use if com and para verificar acesso.

```python
idade = int(input("Digite sua idade: "))
documento = input("Possui documento? (sim/não): ")

if idade >= 18 and documento == "sim":
    print("Acesso liberado!")
else:
    print("Acesso negado!")
            
```

## 5. Peça uma palavra e mostre letra por letra com for.

## 6. Crie um sistema de login com 3 tentativas usando while.

```python
senha_correta = "python123"
tentativas = 0

while tentativas < 3:
    senha = input("Digite a senha: ")

    if senha == senha_correta:
        print("Login realizado com sucesso")
        break
    else:
        tentativas = tentativas + 1
        restantes = 3 - tentativas
        if restantes > 0:
            print(f"Senha incorreta! Você tem {restantes} tentativa(s) restante(s).")

if tentativas == 3:
    print("Conta bloqueada! Número de tentativas esgotado.")
```
