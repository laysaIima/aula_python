# Exercícios com 'If'

### 1. Peça a idade e se a pessoa tem carteira de motorista. Mostre se ela pode dirigir.

```python
idade_pessoa = int(input("Digite sua idade: "))
tem_carteira = input("Tem carteira de motorista? (s/n): ")

if idade_pessoa >= 18 and tem_carteira == 's':
    print("Pode dirigir!")
else:
    print("Não pode dirigir!")
```

### 2. Peça um número. Verifique se ele está entre 50 e 100.

```python
numero = int(input("Digite um número!"))
if numero >=50 and numero <=100:
    print("O número está entre 50 e 100")
else:
    print("O número não está entre 50 e 100")
```

### 3. Peça o dia da semana. Mostre se é final de semana.

```python
dia_semana = input("Digite o dia da semana")
if dia_semana == 'sábado' or dia_semana == 'domingo':
    print("Fim de semana")
else:
    print("Dia útil")
```

### 4. Peça usuário e senha. Verifique se ambos estão corretos.

```python
usuario_correto = "admin"
senha_correta = "54321"
usuario = input("Digite o usuário: ")
senha = input("Digite a senha: ")

if usuario == usuario_correto and senha_correta == "54321":
    print("Usuário e senha corretos, acesso permitido!")
else:
    print("Usuário e senha incorretos, acesso negado!")
```

### 5. Peça uma nota e a frequência do aluno. Mostre se ele foi aprovado.

```python
nota = float(input("Digite a nota (0-100): "))
frequencia = float(input("Digite a frequencia (0-100): "))

if nota >= 70 and frequencia >= 75:
    print("Aluno aprovado!")
else:
    print("Aluno reprovado!")
```

### 6. Peça se o aluno entregou o trabalho. Use not para verificar se ele não entregou.

```python
trabalho_entregue = True
if not trabalho_entregue:
    print("Aluno não entregou o trabalho!")
else:
    print("Aluno entregou o trabalho!")
```

# Exercícios com 'While'

### 1. Mostre os números de 1 até 10 usando while.

```python
contador = 1
while contador <=10:
    print(contador)
    contador = contador + 1
```

### 2. Peça uma senha até o usuário acertar.

```python
senha_correta = "senhacorreta"
senha_digitada = ""

while senha_digitada != senha_correta:
    senha_digitada = input("Digite a senha: ")
    if senha_digitada == senha_correta:
        print("Senha correta! Acesso concedido.")
    else:
        print("Senha incorreta. Tente novamente.")
```

### 3.Peça números até o usuário digitar 0.

```python
numero_correto = "0"
numero_digitado = ""

while numero_digitado != numero_correto:
    numero_digitado = input("Digite um número: ")
    if numero_digitado == numero_correto:
        print("Acertou!")
    else:
        print("Errou! Tente novamente.")
```

### 4. Conte de 10 até 1.

```python
i = 10

while i >= 1:
    print(i)
    i -= 1
```

### 5. Peça uma idade válida até o usuário digitar um valor positivo

```python
idade = int(input("Digite sua idade: "))

while idade <= 0:
    print("Idade inválida! Digite um valor positivo.")
    idade = int(input("Digite sua idade: "))

print(f"Idade válida: {idade} anos")
```

### 6.Crie um menu que só encerra quando o usuário digitar 0.

```python
opcao = -1

while opcao != 0:
    print("─────────────────────")
    print("  1 → Ver saldo")
    print("  2 → Fazer depósito")
    print("  3 → Fazer saque")
    print("  4 → Extrato")
    print("  0 → Sair")
    print("─────────────────────")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        print("Executando: Ver saldo")
    elif opcao == 2:
        print("Executando: Fazer depósito")
    elif opcao == 3:
        print("Executando: Fazer saque")
    elif opcao == 4:
        print("Executando: Extrato")
    elif opcao != 0:
        print("Opção inválida! Tente novamente.")

print("Encerrando o programa... até logo!")
```
