### Exemplo de 'or'

```python
idade_pessoa = int(input("Digite sua idade: "))
estudante = input("Tem documento? (s/n): ")

if idade_pessoa <= 12 or estudante == 's':
    print("Entrada permitida!")
else:
    print("Não pode entrar!")
```

# for
## Exemplo Range()

```python
for numero in range(6):
    print(numero, end=',')
```

```python
# progressão aritimética de razão 3
for numero in range(0, 29, 3):
    print(numero, end=',')
```

```python
# Repetição
for repeticao in range(6):
    print("Estou aprendendo 'for' em python")
```

```python
# sequencia dos 30 primeiros números pares
for numero in range(0, 60, 2):
    print(numero, end=',')
```

```python
# Tabuada do 8
for numero in range(1, 11):
    print(f"8x {numero} =", 8*numero)
```

```python
# Tabuada do 9
for numero in range(1, 11):
    print(f"9x {numero} =", 9*numero)
```

```python
for letra in "Brasil":
    print(letra, end=',')
```

```python
palavra = input("Digite uma palavra: ")

for letra in palavra:
    print(letra, end='-')

letras = len(palavra)
print(f"\n{palavra} tem {letras} letras.")
```
