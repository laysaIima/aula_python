Exercícios Práticos – Nível Fácil

1. Verificar número positivo
Peça ao usuário um número e verifique: Se o número for
maior que 0, mostre "Número positivo".

```python
numero = int(input("Digite um número: "))
print(numero)
if(numero>0):
    print("Número positivo")
```

2. Verificar maior de idade
Peça a idade do usuário. Mostre "Maior de idade" se a
idade for maior ou igual a 18.

```python
idade1 = int(input("Qual a sua idade?"))
print(idade1)
if(idade1>=18):
    print("Maior de idade")
```

3. Número maior que 10
Peça um número ao usuário. Se o número for maior que
10, mostre "Número maior que 10".

```python
numero1 = int(input("Digite um número: "))
print(numero1)
if(numero1>10):
    print("Número maior que 10")
```

4. Par ou Ímpar
Peça um número inteiro. Mostre "Número par" ou
"Número ímpar".

```python
numero2 = int(input("Digite um número: "))
print(numero2)
if(numero2 %2 == 0):
    print("Número par")
else:
    print("Número ímpar")
```

5. Aprovação de aluno
Peça a nota do aluno. Se a nota for maior ou igual a 7,
mostre "Aluno aprovado". Caso contrário, mostre "Aluno
reprovado".

```python
nota = int(input("Digite a nota: "))
print(nota)
if(nota>=7):
    print("Aluno aprovado")
else:
    print("Aluno reprovado")
```

6. Comparar dois números
Peça dois números ao usuário. Mostre qual é maior.
Exemplo: "O maior número é 25".

```python
numero4 = int(input("Digite um número: "))
print(numero4)
numero5 = int(input("Digite um número: "))
print(numero5)
if(numero4>numero5):
    print("O maior número é",numero4)
else:
    print("O maior número é",numero5)
```

7. Classificação de idade
Peça a idade da pessoa. Mostre:
▶ Criança (0–12)
▶ Adolescente (13–17)
▶ Adulto (18 ou mais)

```python
idade = int(input("Qual sua idade?"))
print(idade)
if(idade<13):
    print("Criança")
elif(idade<=17):
    print("Adolescente")
else:
    print("Adulto")
```

8. Nota do aluno
Peça a nota do aluno. Mostre:
▶ Aprovado (nota ≥ 7)
▶ Recuperação (nota entre 5 e 6)
▶ Reprovado (nota < 5)

```python
nota = int(input("Digite a nota: "))
print(nota)
if(nota>=7):
    print("Aprovado")
elif(nota>=5):
    print("Recuperação")
else:
    print("Reprovado")
```

9. Comparar três números
Peça três números. Mostre qual é o maior deles.
Exemplo: "O maior número é 42".

```python
numero1 = int(input("Digite um número: "))
print(numero1)
numero2 = int(input("Digite um número: "))
print(numero2)
numero3 = int(input("Digite um número:"))
print(numero3)
if(numero1>numero2 and numero1>numero3):
    print("O maior número é",numero1)
elif(numero2>numero1 and numero2>numero3):
    print("O maior número é",numero2)
else:
    print("O maior número é",numero3)
```

10. Sistema simples de login
Peça ao usuário nome e senha. Considere:
▶ usuario_correto = "admin"
▶ senha_correta = "1234"
Se os dados estiverem corretos, mostre "Login
realizado com sucesso". Caso contrário, "Usuário ou
senha incorretos".

```python
usuario_correto = "admin"
senha_correta = "1234"
nome = str(input("Digite o usário: "))
senha = str(input("Digite a senha: "))
if(usuario_correto == nome and senha_correta == senha):
    print("Login realizado com sucesso")
else:
    print("Usuário ou senha incorretos.")
```
