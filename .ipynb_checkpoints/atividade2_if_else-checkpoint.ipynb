1. Acesso ao parque
Peça a idade da pessoa. Se ela tiver 12 anos ou mais,
pergunte se está com ingresso (s ou n).
▶ Se tiver idade suficiente e ingresso, mostre: Entrada
liberada
▶ Se tiver idade suficiente e não tiver ingresso, mostre:
Compre o ingresso
▶ Caso contrário, mostre: Idade insuficiente

```python
idade_pessoa = int(input("Digite a idade: "))

if idade_pessoa >= 12:
    tem_ingresso = input("Está com ingresso? (s/n): ")
    if tem_ingresso == "s":
        print("Entrada liberada")
    else:
        print("Compre o ingresso")
else:
    print("Idade insuficiente")
```

2. Login simples
Peça o nome do usuário. Se o usuário for admin, peça a senha.
▶ Se a senha for 1234, mostre: Acesso permitido
▶ Senão, mostre: Senha incorreta
▶ Se o usuário não for admin, mostre: Usuário inválido

```python
nome_usuario = input("Digite o usuário: ")

if nome_usuario == "admin":
    senha_usuario = input("Digite a senha: ")

    if senha_usuario == "1234":
        print("Acesso permitido")
    else:
        print("Senha incorreta")
else:
    print("Usuário inválido")
```

3. Aprovação com presença
Peça a nota do aluno. Se a nota for maior ou igual a 7, peça a
frequência.
▶ Se a frequência for maior ou igual a 75, mostre: Aprovado
▶ Caso contrário, mostre: Reprovado por falta
▶ Se a nota for menor que 7, mostre: Reprovado por nota

```python
nota_aluno = float(input("Digite a nota: "))

if nota_aluno >= 7:
    frequencia_aluno = float(input("Digite a frequência: "))

    if frequencia_aluno >= 75:
        print("Aporvado")
    else:
        print("Reprovado por falta")
else:
    print("Reprovado por nota")
```

4. Número par positivo
Peça um número inteiro.
▶ Se o número for maior que zero, verifique se ele é par
▶ Mostre: Par positivo, Ímpar positivo ou Número não
positivo

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        print("Par positivo")
    else:
        print("Ímpar positivo")
else:
    print("Número não positivo")
```

5. Entrada no cinema
Peça a idade. Se a pessoa tiver 16 anos ou mais, pergunte se
possui documento.
▶ Se possuir documento, mostre: Pode entrar
▶ Caso contrário, mostre: Precisa apresentar documento
▶ Se tiver menos de 16, mostre: Entrada não permitida

```python
idade_pessoa = int(input("Digite sua idade: "))

if idade_pessoa >= 16:
    possui_documento = input("Possui documento? (s/n): " )
    if possui_documento == "s":
        print("Pode entrar")
    else:
        print("Precisa apresentar documento")
else: 
    print("Entrada não permitida")
```

6. Desconto em compra
Peça o valor da compra. Se o valor for maior ou igual a 100,
pergunte se o cliente possui cupom.
▶ Se possuir cupom, mostre: Desconto aplicado
▶ Senão, mostre: Sem desconto
▶ Se a compra for menor que 100, mostre: Valor
insuficiente para desconto

```python
compra = float(input("Qual o valor da compra: "))

if compra >= 100:
    possui_cupom = input("Possui cupom? (s/n): ")
    if possui_cupom == "s":
        print("Desconto aplicado")
    else:
        print("Sem desconto")
else:
    print("Valor insuficiente para desconto")
```

7. Múltiplo de 2 e de 3
Peça um número inteiro.
▶ Se o número for positivo, verifique se ele é múltiplo de 2
▶ Se for múltiplo de 2, verifique também se é múltiplo de 3
▶ Mostre uma mensagem adequada para cada caso

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        if numero % 3 == 0:
            print("O número é positivo, múltiplo de 2 e de 3")
        else:
            print("O número é positivo, e múltiplo de 2, mas não múltiplo de 3")
    else:
        print("O número é positivo, mas não é múltiplo de 2")
else:
    print("O número não é positivo") 
```

8. Cadastro com confirmação
Peça o nome. Se o nome não estiver vazio, peça a idade.
▶ Se a idade for maior ou igual a 18, mostre: Cadastro
completo
▶ Senão, mostre: Cadastro incompleto: menor de idade
▶ Se o nome estiver vazio, mostre: Nome obrigatório

```python
nome = input("Digite seu nome: ")

if nome != "":
    idade = int(input("Digite sua idade: "))
    if idade >= 18:
        print("Cadastro completo")
    else:
        print("Cadastro incompleto: menor de idade")
else:
    print("Nome obrigatório")
```

9. Triagem hospitalar simples
Peça a idade. Se a idade for maior ou igual a 60, pergunte se a
pessoa sente dor forte (s ou n).
▶ Se sim, mostre: Atendimento prioritário imediato
▶ Se não, mostre: Aguardar triagem
▶ Se a idade for menor que 60, mostre: Atendimento comum

```python
idade = int(input("Digite sua idade: "))

if idade >= 60:
    dor_forte = input("Sente dor forte? (s/n): ")

    if dor_forte == "s":
        print("Atendimento prioritário imediato")
    else:
        print("Aguardar triagem")
else:
    print("Atendimento comum")
```

10. Sistema de acesso por níveis
Peça um usuário. Se o usuário for professor, peça a senha.
▶ Se a senha for python123, peça o nível de acesso
▶ Se o nível for 1, mostre: Acesso básico
▶ Caso contrário, mostre: Acesso avançado
▶ Se a senha estiver errada, mostre: Senha inválida
▶ Se o usuário estiver errado, mostre: Usuário não autorizado 

```python
usuario = input("Digite o usuário: ")

if usuario == "professor":
    senha_usuario = input("Digite a senha: ")

    if senha_usuario == "python123":
        nivel_acesso = input("Digite o nível de acesso: ")

        if nivel_acesso == "1":
            print("Acesso básico")
        else:
            print("Acesso avançado")
    else:
        print("Senha inválida")
else:
    print("Usuário não autorizado")
```
